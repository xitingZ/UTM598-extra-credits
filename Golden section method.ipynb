```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = 3

x = np.linspace(a, b)
def f(x):
    return 65*x**6+71*x**5-322*x**4-401*x**3

plt.grid()
plt.plot(x, f(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```

```python
xl = 0.618*a + (1-0.618) * b
xr = (1-0.618)*a + 0.618 * b
zl = 65*xl**6+71*xl**5-322*xl**4-401*xl**3
zr = 65*xr**6+71*xr**5-322*xr**4-401*xr**3

if zl < zr:
    b = xr
else:
    a = xl
        
print(a, b)
x = np.linspace(a, b)
plt.grid()
plt.plot(x, f(x))
plt.xlabel("$x$")
plt.ylabel("$f(x)$")
```
